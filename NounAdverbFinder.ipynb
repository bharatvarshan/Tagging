```python

# spacy is an NLP tool written in python
import spacy
import re
import glob
# we are loading en_core_web_sm into the nlp variable
nlp = spacy.load("en_core_web_sm")
#  path the input text file
path="/content/big.txt"
for file in glob.glob(path):
    with open(file, encoding='utf-8', errors='ignore') as file_in:
        text = file_in.read()
        # Regular expression to match the pattern of words
        words_pattern = '[a-zA-Z]+'
        # find all the words that match the pattern with cases ignored
        lines = re.findall(words_pattern, text, flags=re.IGNORECASE)
        for line in lines:
            line = nlp(line)
            for token in line:
              # for each token spacy.explain provides the noun or verb or adverb etc.
              # if  noun they are printed below
              if spacy.explain(token.pos_)=='noun':
                print("Noun:",token)
              elif spacy.explain(token.pos_)=='verb':
                print("Verb:",token)
```
